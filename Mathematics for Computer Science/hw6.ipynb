```python
import numpy as np
from scipy.optimize import minimize_scalar

#дані
Q = lambda x: 2000 - 0.8*x
P = lambda x: (x - 800) * Q(x)

#максимум прибутку
res = minimize_scalar(lambda x: -P(x), bounds=(800, 2500), method='bounded')

x_opt = res.x
profit_opt = P(x_opt)

print(f"оптимальна ціна = {x_opt:.2f}")
print(f"максимальний прибуток = {profit_opt:.2f}")
```

```python
import numpy as np
from scipy.optimize import minimize

#функція
C = lambda v: v[0]**2 + v[1]**2 - 10*v[0] - 8*v[1] + 50

#мінімізація
res = minimize(C, x0=[0, 0], method='BFGS')

print(f"x* = {res.x[0]:.2f}")
print(f"y* = {res.x[1]:.2f}")
print(f"мінімум = {res.fun:.2f}")
```

```python
import numpy as np
from scipy.optimize import linprog

#коефіцієнти прибутку
c = [-500, -800]  #мінус бо linprog мінімізує

#обмеження
A = [[2, 4],
     [3, 2]]
b = [120, 90]

#умови невідʼємності
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A, b_ub=b, bounds=bounds)

x, y = res.x
profit = 500*x + 800*y

print(f"стільці = {x:.2f}")
print(f"столи = {y:.2f}")
print(f"прибуток = {profit:.2f}")
```

```python
import numpy as np
from scipy.optimize import linprog

#цільова функція (мінус, бо мінімізуємо)
c = [-500, -800]

#обмеження
A = [[2, 4],
     [3, 2]]
b = [120, 90]

#межі
bounds = [(0, None), (0, None)]

res = linprog(c, A_ub=A, b_ub=b, bounds=bounds)

x, y = res.x

print(f"x = {x:.2f}")
print(f"y = {y:.2f}")
```

```python
#використані ресурси
wood = 2*x + 4*y
time = 3*x + 2*y

print(f"деревина = {wood:.2f}")
print(f"робочий час = {time:.2f}")
```
