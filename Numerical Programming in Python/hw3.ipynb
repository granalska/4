```julia
import pandas as pd #бібліотека для датафрейм
import numpy as np 
import pickle
import os
from sklearn.decomposition import PCA

#if os.path.exists('/content'):
    #для колаб
    #path = 'word_embeddings_subset.p'
#else:
    #через робочий стіл
path = '/Users/admin/Desktop/word_embeddings_subset.p'

with open(path, 'rb') as f:
    word_embeddings = pickle.load(f)

words = list(word_embeddings.keys())
vectors = list(word_embeddings.value())

pca = PCA(n_components=3) #тривимірні вектори для всіх слів
embeddings3d = pca.fit_transform(vectors)
print(np.shape(embeddings3d)) 


df_3d = pd.DataFrame(embeddings3d) #інф про слова та їхні тривимірні вектори
df_3d ['word'] = words
df_3d.head()
```
